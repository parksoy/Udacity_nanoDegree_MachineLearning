```python
import idx2numpy
import numpy as np

# read data and convert idx file to numpy array
ndarr_train = idx2numpy.convert_from_file('train-images-idx3-ubyte.idx')
labels_raw_train = idx2numpy.convert_from_file('train-labels-idx1-ubyte.idx')
ndarr_test = idx2numpy.convert_from_file('t10k-images-idx3-ubyte.idx')
labels_raw_test = idx2numpy.convert_from_file('t10k-labels-idx1-ubyte.idx')


image_height = 28
image_width = 140 # 28*5 
num_digit=5

def createSequences(ndarr,labels_raw):
    
    dataset_size = ndarr.shape[0]/5
    dataset=[]
    data_labels = []
    
    dataset = np.ndarray(shape=(dataset_size, image_height, image_width),dtype=np.float32)
    data_labels=np.ndarray(shape=(dataset_size,num_digit+1),dtype=np.int32)

    i = 0
    w = 0
    while i < dataset_size: #dataset_size
        temp = np.hstack(
            [ndarr[w], ndarr[w + 1], ndarr[w + 2], ndarr[w + 3], ndarr[w + 4]])
        
        dataset[i, :, :] = temp
        
        #need to make train_labels.shape=(# of image files, 6)
        temp_str = np.hstack([labels_raw[w], labels_raw[w + 1], labels_raw[w + 2], labels_raw[w + 3], labels_raw[w + 4], 
                    num_digit])
        data_labels[i,:]=temp_str #.append(temp_str)
        
        
        w += 5
        i += 1

    np.array(data_labels)

    return dataset, data_labels



concat_dataset_train, concat_data_labels_train = createSequences(ndarr_train,labels_raw_train)
concat_dataset_test, concat_data_labels_test = createSequences(ndarr_test,labels_raw_test)
```

```python
from PIL import Image

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (5.0, 5.0)

img=Image.fromarray(concat_dataset_train[0])
img2=Image.fromarray(concat_dataset_train[1])


longer_side = max(img.size) #(252, 143)
horizontal_padding = (longer_side - img.size[0]) / 2
vertical_padding = (longer_side - img.size[1]) / 2
img2 = img.crop(
    (
        -horizontal_padding,
        -vertical_padding,
        img.size[0] + horizontal_padding,
        img.size[1] + vertical_padding
    )
)


'''
'''
f, ax = plt.subplots(nrows=5, ncols=1)
#print(f)
#print(ax)


ax[0].set_title('img', loc='center')
print('img.size==',img.size)
ax[0].imshow(img)

ax[1].set_title('img2', loc='center')
ax[1].axis('off')
print('img2.size==',img2.size)
ax[1].imshow(img2)


img2_resized_from_140_140=img2.resize((28,28),Image.ANTIALIAS)

ax[2].set_title('img2_resized_from_140_140', loc='center')
ax[2].axis('off')
print('img2_resized_from_140_140.size=',img2_resized_from_140_140.size)
ax[2].imshow(img2_resized_from_140_140)


img3_resized_from_28_140=img.resize((28,28),Image.ANTIALIAS)
ax[3].set_title('img3_resized_from_28_140', loc='center')
ax[3].axis('off')
print('img3_resized_from_28_140.size=',img3_resized_from_28_140.size)
ax[3].imshow(img3_resized_from_28_140)

from PIL import ImageFilter
from PIL import ImageEnhance


#img4=img3_resized_from_28_140.filter(ImageFilter.SHARPEN)
#img4=ImageEnhance.Sharpness(img3_resized_from_28_140)

#ax[4].imshow(img4)



''''''
```
