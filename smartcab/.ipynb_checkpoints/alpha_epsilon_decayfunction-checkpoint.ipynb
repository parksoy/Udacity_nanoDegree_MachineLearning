```python
#Things to optimize
tolerance_init=0.05 # epsilon tolerance before beginning testing, default is 0.05
a_init=0.5 # learning rate,alpha,default 0.5
e_init=1 #exploration factor,default 1

#e=[0.95, 0.8999999999999999, 0.8499999999999999, 0.7999999999999998, 0.7499999999999998, 0.6999999999999997, 0.6499999999999997, 0.5999999999999996, 0.5499999999999996, 0.4999999999999996, 0.4499999999999996, 0.39999999999999963, 0.34999999999999964, 0.29999999999999966, 0.24999999999999967, 0.19999999999999968, 0.1499999999999997, 0.09999999999999969, 0.049999999999999684, -3.191891195797325e-16]
```


$$ \epsilon_{t+1} = \epsilon_{t} - 0.05, \hspace{10px}\textrm{for trial number } t$$


$$ \epsilon = a^t, \textrm{for } 0 < a < 1 \hspace{50px}\epsilon = \frac{1}{t^2}\hspace{50px}\epsilon = e^{-at}, \textrm{for } 0 < a < 1 \hspace{50px} \epsilon = \cos(at), \textrm{for } 0 < a < 1$$

```python
#decay function
import math
import numpy as np


t=range(0,21) #step in one training trial
t_np=np.array(t)
print "t_np==",t_np

#linear
e_np_lin=e_init-t_np*0.05

#power
e_np_pow=a_init**t_np
#print "e_np_pow==",e_np_pow

#denominal
#print "t_np^2==",t_np**2
e_np_den=1/(t_np**2)

#exponential
at=a_init*t_np
e_np_exp=np.exp(-1*at)

#cos
e_np_cos=np.cos(at)
print "e_np_cos==",

```

```python
import pylab as pl
import matplotlib.pyplot as plt
%matplotlib inline

def plotting():
    lin,=plt.plot(t_np,e_np_lin,'go-')
    exp,=plt.plot(t_np,e_np_exp,'bo-')
    den,=plt.plot(t_np,e_np_den,'ro-')
    powe,=plt.plot(t_np,e_np_pow,'ko-')
    cosi,=plt.plot(t_np,e_np_cos,'bx-')
    plt.title('Plot of e vs. t when a={}, e={}'.format(a_init,e_init))    #"{:10.4f}".format(x)
    plt.xlabel('t')
    plt.ylabel('e')
  
    plt.legend([lin, exp, den,powe,cosi], ['linear', 'exponential','denominator','power','cosine'],fontsize=10)
    
    return 

plotting()
```

```python
a_init=[0.9,0.7,0.5,0.3] # learning rate,alpha,default 0.5
e_init=[0.5,0.7,0.9,1] #exploration factor,default 1

def plotting_looping_a_e():
    lin,=plt.plot(t_np,e_np_lin,'go-')
    exp,=plt.plot(t_np,e_np_exp,'bo-')
    
    plt.title('Plot of e vs. t when a={}, e={}'.format(a_init,e_init))    
    plt.xlabel('t')
    plt.ylabel('e')

    plt.legend([lin, exp], ['linear', 'exponential'],fontsize=10) #, a={}, e={},, format(a_init,e_init)
    
    return 

for a in a_init:  
    for e in a_init:
        #linear
        e_np_lin=e-t_np*0.05
        #print "e_np_lin--",e_np_lin

        #exponential
        at=a*t_np
        e_np_exp=np.exp(-1*at)
        #print "e_np_exp--",e_np_exp
        
        plotting_looping_a_e()
        
```
